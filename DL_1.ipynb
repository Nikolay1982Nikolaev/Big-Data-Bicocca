## The backpropagation algorithm:

1. Intialize weights
2. compute the activations for each leyer the first to the last (forward pass)

3. compute the classification error for the m instances through a loss function, for instance:
$$E(w)= \sum_{m=1}^M (y_m - f(x_m|w,b))^2$$

3. Compute the gradient of error function $E(w)$ wrt w (backwar pass - iterative since for every layer w and b depend on the previous layer)

5. Updata the weights based on the following rule:
$$w^{k+1}= w^k - \eta \frac{\partial E(w)}{\partial w}|_{w^k}$$
- $\eta$ learning rate, k iterations

## Why derivatives?

$$a= x_1*w_1 + x_2*w_2- b$$

- if $w_2$ increase for some value $\eta$ then it would increase a by $5\eta$
- remember that w can also be negative

$\frac{\partial E(w)}{\partial w}$ tells us the sensitivity of the error with respect to each direction (weight). Thus, we can modify the weights accordingly -> $s^{k+1}= w^k - \eta \frac{\partial E(w)}{\partial w}|_{w^{k+1}}$ $\eta$ learning rate, k interations

## Backpropagation in perceptron:

$$E= \frac{1}{2}(y-1)^2 = \frac{1}{2}(y-\sigma(\sum w_i x_i))^2$$

$$\frac{\partial E}{\partial w_1}= \sum_j \frac{\partial E}{\partial a}\frac{\partial a}{\partial z}\frac{\partial z}{\partial w_1}$$ - j-samples

For 1 sample and $w_1$:

$$\frac{\partial a}{\partial z}= \sigma(z)(1-\sigma(z))$$

## Activation function:

sometime adding non-linearity is good:

$$sigmoid: \sigma(z)= \frac{1}{1+e^{-z}}$$

$$tanh: \sigma(z)= \frac{e^z- e^{-z}}{e^z + e^{-z}}$$



## The vanishing /exploding gradient:

Activations such as sigmoid and tanh satures:
- far from 0, gradient is close to 0
- backpropagation requires gradient multiplication
- gradient far from the output vanishes
- learning doesn't happen

- close to 0, gradient canb be >>1
- backpropagation requires gradient multiplication
- gradient far from the output explodes
- learning doesn't happen

Recfigied Linear Unit:
$$ReLU (a)= max(0, a)$$
+ (+) faster SGD convergence
+ (+) No vanishing and explofing gradient
+ (+) scale - invariant: max(0, ba)= b(max(0,a))
